```python
#P3

def estimate_pi(N):
    points = np.random.rand(N, 2)
    distances = np.linalg.norm(points, axis=1)
    points_inside = np.sum(distances <= 1)
    pi_estimate = 4 * points_inside / N
    return pi_estimate, points

N_values = [int(1e3), int(1e4), int(1e5), int(1e6)]
pi_estimates = []
for N in N_values:
    pi_estimate, points = estimate_pi(N)
    pi_estimates.append(pi_estimate)
    print(f"N = {N}: Estimated pi = {pi_estimate}")

plt.figure(figsize=(6, 6))
plt.axes().set_aspect('equal', 'box')
plt.title('for N = 1e4')
plt.xlabel('x')
plt.ylabel('y')

inside_circle = points[np.linalg.norm(points, axis=1) <= 1]
outside_circle = points[np.linalg.norm(points, axis=1) > 1]
plt.scatter(inside_circle[:, 0], inside_circle[:, 1], color='blue', label='Inside Circle')
plt.scatter(outside_circle[:, 0], outside_circle[:, 1], color='red', label='Outside Circle')

theta = np.linspace(0, np.pi/2, 100)
plt.plot(np.cos(theta), np.sin(theta), color='black')
plt.text(0.1, 0.9, f'  {pi_estimates[1]}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
plt.legend()
plt.show()
```
